Now can I take this all into a function?

```python
def race_victories(year, teamName):
    championship = results[results.year == year] #get the right set of races
    team = championship[championship.constructorName == teamName] #get the results for the team in question
    team_wins = team[championship.positionText == "1"] #get all the victories of the team
    
    rounds = championship["round"].max() #how many total races in the season
    
    victories = list(team_wins.prixName) #the list of all races won
    rounds_won = len(victories) #number of races won
    
    dominance = rounds_won / float(rounds)
    return [year, teamName, dominance, rounds, rounds_won, victories]
```

```python
race_victories(1950, "Alfa Romeo")
```

From this I can feed it an array of year,team combinations and get the info I want.

```python
import csv
```

```python
with open("../data/teams.csv", "r") as f:
    teams = csv.reader(f)
    constructors = list(teams)
```

```python
test = [["year", "team", "dominance %", "rounds", "races won",  "victories"]]
for i in range(1,len(constructors)):
    year = int(constructors[i][0])
    team = str(constructors[i][1])
    line = race_victories(year, team)
    test.append(line)
```

```python
constructors[0]
```

```python
test[0:3]
```

```python
df = pd.DataFrame(np.array(test[1:]), columns= test[0])
```

That took a while, but let's redo everything with `teams2.csv`

```python
with open("../data/teams2.csv", "r") as f:
    teams = csv.reader(f)
    constructors = list(teams)
    
summary = [["year", "team", "dominance %", "rounds", "races won",  "victories"]]
for i in range(1,len(constructors)):
    year = int(constructors[i][0])
    team = str(constructors[i][1])
    line = race_victories(year, team)
    summary.append(line)
    
df = pd.DataFrame(np.array(summary[1:]), columns= summary[0])
```
